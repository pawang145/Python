
Q1:- Write a program in Python to prompt the user to input their name, employee number, week ending date, hours worked, rate per hour, standard and overtime tax percentage rate. Use the data input to calculate gross pay, tax deductions and net pay. Output the results as a formatted payslip. Assume that a standard working week is 37.5 hours.


```
#### Dublin Business School ####

### Program For Payslip#
import datetime
standerdweeklyhours = 37.50
OvertimeRate = 1.5

#Inputs From User
EmpName = input("Please type Emp name: ")
Empnumber = input("Please type in your Emp number: ")
Numberoffworkinghours = float(input("Please type  Number Of Hours Worked:  "))
Hrworked = float(input("Please type  Hourly Rate: "))

#calculation Of Tax
Stdworked = int(input("Please type Standard Tax Rate: "))
Stdworked = Stdworked / 100.0
Totalnormalpay = standerdweeklyhours * Hrworked
Regulartax = Totalnormalpay * Stdworked
OvertimeTaxRate = int(input("Please input Overtime Tax Rate: "))
OvertimeTaxRate = OvertimeTaxRate / 100
overtimetaxrate = OvertimeRate * Hrworked

#total pay 
Totalnormalpay = standerdweeklyhours * Hrworked

#For Different format it will print wrong format & again take the input from user
while True:
    try:
        date = input("Enter date in YYYY/MM/DD format: ")
        day, month, year = map(int, date.split('/'))
        WeekEnding = datetime.date(day, month, year)
        break

    except:
        print("You entered the wrong format")

#calculation Part for Overtime
if Numberoffworkinghours > standerdweeklyhours:               #     
   HoursOverTime = Numberoffworkinghours - standerdweeklyhours#
   TotalOverTime = HoursOverTime * overtimetaxrate            #
   OverTimeTax = OvertimeTaxRate * TotalOverTime              #
   TotalPay = Totalnormalpay + TotalOverTime                  #
   TotalDeduction = Regulartax + OverTimeTax                  #
   NetPay = TotalPay - TotalDeduction                         #

#caluation Part for standerd time
else:
    HoursOverTime = 0
    TotalOverTime = HoursOverTime * overtimetaxrate           #
    OverTimeTax = OvertimeTaxRate * TotalOverTime             #
    TotalPay = Totalnormalpay + TotalOverTime                 #
    TotalDeduction = Regulartax + OverTimeTax                 #
    NetPay = TotalPay - TotalDeduction                        #

print(" ")
print("\t                                        PAYSLIP")
print(' ')
print("WEEK ENDING: ",WeekEnding.strftime("%d/%m/%Y"))
print("Employee: ", EmpName)
print("Employee Number: ",Empnumber)
print(" ")
print("                                Earnings                 Deductions")
print("                                Hours    Rate    Total")
print("Hours (normal)                  %.2f"%standerdweeklyhours, "\t","%.2f"%Hrworked,"\t","%.2f"%Totalnormalpay,"  Tax  @","{:.0%}".format(Stdworked), Regulartax)
print("Hours (overtime)                %.2f"%HoursOverTime,"\t","%.2f"%overtimetaxrate,"\t","%.2f"%TotalOverTime,"  Tax  @","{:.0%}".format(OvertimeTaxRate),OverTimeTax,"\n","\n")
print("Total pay:","          ","%.2f"%TotalPay)
print("Total deductions:","   ","%.2f"%TotalDeduction)
print("Net pay:","            ","%.2f"%NetPay)
```

Q4:- Implementation of phonebook 

```
#######phoneBook using Python######
class PhoneBook():

  ### assigning the variable for operation ###

    def __init__(self):
        self.phonebk = {}
        operation = input("Which operation you want to perform (ADD, SEARCH, UPDATE & DELETE)\n")
        if operation.lower() == "addition" or operation.lower() == "a":
            self.addContact()
        elif operation.lower() == "search" or operation.lower() == "s":
            self.searchContact()
        elif operation.lower() == "update" or operation.lower() == "u":
            self.updateContact()
        elif operation.lower() == "delete" or operation.lower() == "d":
            self.deleteContact()
        else:
            print("Enter valid Choice")
            self.againOperation()

    def againOperation(self):
        request = input("do you want to perform another operation: YES or NO ?\n")
        if request.lower() == "yes" or request.lower() == "y":
            self.operation = input("choose the operation you want to perform (ADD, SEARCH, UPDATE & DELETE)\n")
            if self.operation.lower() == "add" or self.operation.lower() == "a":
                self.addContact()
            elif self.operation.lower() == "search" or self.operation.lower() == "s":
                self.searchContact()
            elif self.operation.lower() == "update" or self.operation.lower() == "u":
                self.updateContact()
            elif self.operation.lower() == "delete" or self.operation.lower() == "d":
                self.deleteContact()
            else:
                print("Enter valid Choice")
                self.againOperation()
        else:
            print("Thank you for using this PhoneBook")

    def addContact(self):
        name = input("Please enter the Name\n")
        if name.title() in self.phonebk.keys():
            print("Contact already exists")
            self.againOperation()
        else:
            contactNum = input("Please enter Contact Number\n")
            if contactNum.isnumeric() == True and contactNum not in self.phonebk.values() and len(contactNum)==10:
                self.phonebk.update({name.title():contactNum})
                print("Contact added successfully")
                self.againOperation()
            elif contactNum.isnumeric() == True and contactNum in self.phonebk.values():
                print("Contact number already associated with other contact")
                self.againOperation()
            else:
                print("Enter a valid 10 digit Number")
                self.againOperation()

    def searchContact(self):
        search = input("Please enter the Name / Number to search\n")
        for i,j in self.phonebk.items():
            if search.title() == i:
                print(i,j)
                self.againOperation()
            if search.title() == j:
                print(i,j)
                self.againOperation()


    def updateContact(self):
        update = input("Please enter the Name / Number to update\n")
        for i,j in self.phonebk.items():
            if update.title() == i:
                new_name = input("Please enter updated name\n")
                self.phonebk.update({new_name:j})
                del (self.phonebk[i])
                print(new_name,j)
                print(self.phonebk)
                self.againOperation()
            elif update.title() == j:
                new_num = input("Please enter updated number\n")
                self.phonebk.update({i:new_num})
                print(i,new_num)
                print(self.phonebk)
                self.againOperation()
        else:
                print("Contact not found")
                self.againOperation()

    
    def deleteContact(self):
        delete = input("Please enter the name / number to delete\n")
        if len(self.phonebk) == 0:
            print("No Contacts in Phonebook")
            self.againOperation()
        else:
            for i,j in self.phonebk.items():
                if delete.title() == i:
                    del(self.phonebk[i])
                    print("Contact deleted successfully")
                    self.againOperation()
                elif delete.title() == j:
                    del(self.phonebk[i])
                    print("Number deleted successfully")
                    self.againOperation()

object = PhoneBook()
```
